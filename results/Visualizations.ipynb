```python
df = pd.read_csv(r"C:\Users\hidoz\Downloads\Content Based Filtering Recommender Systems\data\clean_news_with_text.csv")
```

```python
plt.figure(figsize=(12,6))
top_categories = df['category'].value_counts().head(10)
sns.barplot(x=top_categories.values, y=top_categories.index, palette="viridis")
plt.title("Top 10 News Categories")
plt.xlabel("Number of Articles")
plt.ylabel("Category")
plt.tight_layout()
plt.savefig("results/category_distribution.png")
plt.show()
```

```python
text_data = ' '.join(df['text'].dropna().values[:5000])  # limit sample
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_data)

plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of News Articles")
plt.savefig("results/word_cloud.png")
plt.show()
```
