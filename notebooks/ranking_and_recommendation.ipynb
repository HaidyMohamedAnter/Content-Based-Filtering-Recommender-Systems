```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
os.makedirs("results", exist_ok=True)
```

```python
recommendations = pd.read_csv("results/sample_recommendations.csv")
```

```python
for i, row in recommendations.iterrows():
    print(f"{i+1}. {row['title']} (Similarity: {row['similarity']:.3f})")
```

```python
titles = recommendations['title']
similarities = recommendations['similarity']
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=similarities, y=titles, palette='coolwarm')
```
